```python
# imports
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#load the data files 
basics = pd.read_csv("zippedData/imdb.title.basics.csv.gz")
ratings = pd.read_csv("zippedData/imdb.title.ratings.csv.gz")
gross = pd.read_csv("zippedData/bom.movie_gross.csv.gz")
budgets = pd.read_csv("zippedData/tn.movie_budgets.csv.gz")
```

```python
#Clean title basics dataframe
basics.columns #check what column names are there in basics data frame
```

```python
basics.head() # shows what data is in each column
```

```python
basics.info
```

```python
basics.replace('\\N',pd.NA, inplace=True) #imdb writes \N when data is missing and pandas can't understand that \N
#the pd.Na shows the value missing and the inplace stops creation of new data Frames 

basics.drop_duplicates(inplace=True) # removes duplicates of data
```

```python
basics['primary_title'] = basics['primary_title'].str.strip() #removes spaces before and after text
basics['primary_title'] = basics['primary_title'].str.lower() #lowercases all letters 
```

```python
basics['original_title'] = basics['original_title'].str.strip()
basics['original_title'] = basics['original_title'].str.lower()
```

```python
basics['runtime_minutes'] = pd.to_numeric(basics['runtime_minutes'], errors='coerce') # converts the texts into numbers the errors=coerce makes the whole column numeric
basics['start_year'] = pd.to_numeric(basics['start_year'], errors='coerce')
basics['start_year'] = basics['start_year'].fillna(0, inplace = True) #fills the missing values at the start year column with 0
```

```python
median_runtime = basics['runtime_minutes'].median() #calculates the median of the runtime column
basics['runtime_minutes'] = basics['runtime_minutes'].fillna(median_runtime) #fills the missing values (Na) with the median value
```

```python
basics['filmType'] = basics['runtime_minutes'].apply(lambda x : 'short' if x < 40 else 'movie')
```

```python
basics["genres"] = basics["genres"].apply(lambda x: x.split(",") if type(x) == str else x) 
# splits genres into list while leaving the missing values as they are 
```

```python
# clean ratings dataframes
ratings.columns #checks what the column names are 
```

```python
ratings.head() # checks what the values of each column are
```

```python
ratings.info()
```

```python
ratings.drop_duplicates(inplace=True) #remove duplicates 
```

```python
ratings['averagerating'] = pd.to_numeric(ratings['averagerating'], errors='coerce') #changing the texts to numeric numbers
ratings['numvotes'] = pd.to_numeric(ratings['numvotes'], errors='coerce')

ratings['averagerating'] = ratings['averagerating'].fillna(ratings['averagerating'].median()) #fills the missing values with the median
ratings['numvotes'] = ratings['numvotes'].fillna(0) #fills the missing values in the numvotes with 0
```

```python
ratings = ratings[ratings['averagerating'].between(0, 10)] #have ratings between 0 and 10
```
